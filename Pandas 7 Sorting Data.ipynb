```python
people = {
    'first' : ['Alex', 'Donald', 'Phillipa', 'Jane', 'John', 'Alan'],
    'last' : ['McIntosh', 'Smith', 'Jones', 'Baker', 'Doe', 'Doe'],
    'email' : ['alexmc124@gmail.com','donald88@aol.com','philjones@gmail.com','janebaker@aol.com', 'JohnDoe@email.com', 'AlanDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#Sorting 1 Column using sort_values 
df.sort_values(by='last')
```

```python
# can also be used to sort ascending 
df.sort_values(by='last', ascending=False)
```

```python
# can also sort by one condition then another 
df.sort_values(by=['last','first'], ascending=False)
```

```python
# can also sort by one condition then another BOTH Asc then Dec  
df.sort_values(by=['last','first'], ascending=[False,True])
```

```python
# Can Also sort by index 
df.sort_index()
```

```python
#Can also sort by series 
df['last'].sort_values()
```
