```python
people = {
    'first' : ['Alex', 'Donald', 'Phillipa', 'Jane', 'John', 'Alan'],
    'last' : ['McIntosh', 'Smith', 'Jones', 'Baker', 'Doe', 'Doe'],
    'email' : ['alexmc124@gmail.com','donald88@aol.com','philjones@gmail.com','janebaker@aol.com', 'JohnDoe@email.com', 'AlanDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#combine the first and last name 
df['first'] + ' ' +  df['last']
```

```python
# Assign to the new Column
df['full_name'] = df['first'] + ' ' +  df['last']
```

```python
df
```

```python
df['first']
```

```python
# Removing a Column 
df.drop(columns=['first','last'], inplace=True)
```

```python
df
```

```python
# Splitting a Column with EXPAND
df['full_name'].str.split(' ', expand=True)
```

```python
#Assigning the Lists to the Data frame 
df[['first','last']] = df['full_name'].str.split(' ', expand=True)
```

```python
df
```

```python
# Add a single row to the bottom using APPEND 
df.append({'first': 'Tony'}, ignore_index = True)
```

```python
people = {
    'first' : ['Alex', 'Donald'],
    'last' : ['McIntosh', 'Smith'],
    'email' : ['alexmc124@gmail.com','donald88@aol.com']
}
df2 = pd.DataFrame(people)
```

```python
#Adding the Values of df2 to DF 1 
df.append(df2, ignore_index=True)
```

```python
df = df.append(df2, ignore_index=True)
```

```python
# Removing a Row using the index
df.drop(index=4)
```

```python
 # Removing Rows by conditioning 
df.drop(index=df[df['last'] == 'Doe'].index)
```

```python
# Can also be done using  a filter before hand
filt = df['last'] == 'Doe'
df.drop(index=df[filt].index)
```
